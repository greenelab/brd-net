```python
import pandas
from plotnine import *
```

```python
result_df = pandas.read_csv('../results/model_eval_results.csv', index_col=0)

result_df['acc_over_baseline'] = result_df['val_acc'] - result_df['val_baseline']

result_df.head()
```

```python
ggplot(result_df, aes(x='Model', y='val_acc')) + geom_point()
```

```python
ggplot(result_df, aes(x='Model', y='acc_over_baseline')) + geom_point()
```
